# Matplotlib

```python
import numpy as np
```

```python
from  matplotlib import pyplot as plt
```

```python
from matplotlib import style #adding style
```

# Linear Graph

```python
x=np.array([5,6,7])
y=np.array([1,2,4])
plt.plot(x,y,label="linear graph",color='g')
plt.title("Sample Linear Graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# Scatter Plot

```python
x=np.array([5,6,7])
y=np.array([1,2,4])
plt.scatter(x,y,label="Scatter plot",color='m', marker="X")
plt.title("Sample Scatter Graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True,color='k')
plt.legend()
plt.show()
```
